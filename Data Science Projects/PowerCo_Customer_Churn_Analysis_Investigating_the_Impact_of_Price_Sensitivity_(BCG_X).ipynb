## Data visualization

```python
def plot_stacked_bars(dataframe, title_, size_=(18, 10), rot_=0, legend_="upper right"):
    """
    Plot stacked bars with annotations
    """
    ax = dataframe.plot(
        kind="bar",
        stacked=True,
        figsize=size_,
        rot=rot_,
        title=title_
    )

    # Annotate bars
    annotate_stacked_bars(ax, textsize=14)
    # Rename legend
    plt.legend(["Retention", "Churn"], loc=legend_)
    # Labels
    plt.ylabel("Company base (%)")
    plt.show()

def annotate_stacked_bars(ax, pad=0.99, colour="white", textsize=13):
    """
    Add value annotations to the bars
    """

    # Iterate over the plotted rectanges/bars
    for p in ax.patches:

        # Calculate annotation
        value = str(round(p.get_height(),1))
        # If value is 0 do not annotate
        if value == '0.0':
            continue
        ax.annotate(
            value,
            ((p.get_x()+ p.get_width()/2)*pad-0.05, (p.get_y()+p.get_height()/2)*pad),
            color=colour,
            size=textsize
        )

def plot_distribution(dataframe, column, ax, bins_=50):
    """
    Plot variable distirbution in a stacked histogram of churned or retained company
    """
    # Create a temporal dataframe with the data to be plot
    temp = pd.DataFrame({"Retention": dataframe[dataframe["churn"]==0][column],
    "Churn":dataframe[dataframe["churn"]==1][column]})
    # Plot the histogram
    temp[["Retention","Churn"]].plot(kind='hist', bins=bins_, ax=ax, stacked=True)
    # X-axis label
    ax.set_xlabel(column)
    # Change the x-axis to plain style
    ax.ticklabel_format(style='plain', axis='x')
```

Thhe first function `plot_stacked_bars` is used to plot a stacked bar chart. An example of how you could use this is shown below:

```python
churn = client_df[['id', 'churn']]
churn.columns = ['Companies', 'churn']
churn_total = churn.groupby(churn['churn']).count()
churn_percentage = churn_total / churn_total.sum() * 100
plot_stacked_bars(churn_percentage.transpose(), "Churning status", (5, 5), legend_="lower right")
```

The second function `annotate_bars` is used by the first function, but the third function `plot_distribution` helps you to plot the distribution of a numeric column. An example of how it can be used is given below:

```python
consumption = client_df[['id', 'cons_12m', 'cons_gas_12m', 'cons_last_month', 'imp_cons', 'has_gas', 'churn']]

fig, axs = plt.subplots(nrows=1, figsize=(18, 5))

plot_distribution(consumption, 'cons_12m', axs)
```

**Insight Summary**

1. **Price is a critical factor in churn:**
Analysis has shown that highly price-sensitive customers are the most vulnerable to churn, especially in the absence of personalized offers or flexible plans.

2. **Usage patterns are directly related to churn:**
Customers who use the service infrequently or with low churn are more likely to churn than others, indicating an opportunity to redesign the value proposition for this segment.

3. **Specific demographic characteristics predict higher churn rates:**
Certain geographic regions and certain age groups exhibited higher-than-average churn rates, necessitating precise targeting in marketing campaigns and communications.

**Recommendations**

1. **Design flexible and targeted pricing packages:**
Offer customized packages to the most price-sensitive segments, while adopting dynamic pricing strategies that respond to changes in the market and customer behavior.

2. **Adopting an early warning system for churn-prone customers:**
Run a predictive model periodically to identify customers at risk of churn and automatically connect them with offers or proactive personalized communication.

3. **Improving the digital experience:**
Developing user interfaces on digital platforms and providing smart consumption alerts and more interactive communication options.

4. **Launching geographic and demographic-targeted marketing campaigns:**
Tailoring marketing messages and offers based on customer demographics to maximize loyalty.

5. **Linking customer satisfaction indicators to team performance:**
Regularly implementing tools such as the NPS index, and using its results as a basis for improving service quality and measuring internal performance.
